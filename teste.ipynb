```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("avengers.csv", encoding="iso-8859-1")
df.head(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Return1'] = df['Return1'].fillna("NO")
```

```python
df['Return1'] = df['Return1'].map({'YES': 1, 'NO': 0})
df['Death1'] = df['Death1'].map({'YES': 1, 'NO': 0})
df.head(5)
```

```python
# Qual a quantidade de herois?
qtd_herois = df['Name/Alias'].nunique()
qtd_herois
```

```python
# Qual a distribuiÃ§Ã£o de gÃªnero dos personagens? 
contagem = df['Gender'].value_counts()
contagem.plot.pie(
    autopct='%1.1f%%',   # mostra a porcentagem em cada fatia
    startangle=90,       # gira o grÃ¡fico para ficar bonito
    colors=['lightblue', 'lightpink'],  # cores opcionais
    legend=True          # mostra legenda
)

plt.title("DistribuiÃ§Ã£o de GÃªneros dos Vingadores")  # tÃ­tulo do grÃ¡fico
plt.ylabel('')  # remove o label do eixo y
plt.show()      # mostra o grÃ¡fico
```

```python
# DistribuiÃ§Ã£o da coluna Appearances - variavel numerica - histograma 
plt.figure(figsize=(10,6))
plt.hist(df['Appearances'], bins=10, color='skyblue', edgecolor='black')
plt.title('DistribuiÃ§Ã£o de ApariÃ§Ãµes dos Personagens')
plt.xlabel('NÃºmero de ApariÃ§Ãµes')
plt.ylabel('Quantidade de Personagens')
plt.show()
```

```python
# # DistribuiÃ§Ã£o da coluna Appearances - variavel numerica - boxplot
plt.figure(figsize=(8,6))
plt.boxplot(df['Appearances'], vert=True)
plt.title('Boxplot das ApariÃ§Ãµes dos Personagens')
plt.ylabel('NÃºmero de ApariÃ§Ãµes')
plt.show()
```

```python
# DistribuiÃ§Ã£o da coluna Appearances - variavel numerica - grafico violino
plt.figure(figsize=(8,6))
sns.violinplot(y=df['Appearances'], color='lightgreen')
plt.title('Violin Plot das ApariÃ§Ãµes dos Personagens')
plt.ylabel('NÃºmero de ApariÃ§Ãµes')
plt.show()
```
